**0_1. 基本的寫入與讀取**(使用List)

```
from bs4 import BeautifulSoup as bs
from google.colab import files
import requests, csv

year = []
person = []

url = "https://www.daxi-hro.tycg.gov.tw/home.jsp?id=25&parentpath=0,21,22"

content = requests.get(url).text
parse = bs(content, "html.parser")

data1 = parse.find_all("tbody")
data1 = data1[1]
rows = data1.find_all("tr")
for row in rows:
    cols = row.find_all("td")
    #print(cols)
    if len(cols) > 0 and cols[1].text != '─':
        str1 = cols[0].text
        str1 = str1[0:-1]
        year.append(int(str1))
        #print(year)
        person.append(int(cols[1].text))
        #print(person)

print("桃園市大溪區歷年戶數")
print("=================")

#i = len(year)-1
#while i > -1:
    #print("%3d年: %d戶"%(year[i],person[i]))
    #i -= 1

# 建立csv檔案
with open('桃園市大溪區歷年戶數.csv','w',newline='',encoding='big5') as f :
  # 建立 CSV 檔寫入器
  writer = csv.writer(f)
  # 寫入第一列資料
  writer.writerow(['年分','戶數'])
  for j in range(len(year)):
    writer.writerow([year[j],person[j]])

f.close()

# 下載檔案
files.download('桃園市大溪區歷年戶數.csv') 

# 讀取csv檔案並列印
with open('桃園市大溪區歷年戶數.csv','r',newline='',encoding='big5') as f :
  # 建立 CSV 檔讀入器
  reader = csv.reader(f)
  for j in reader:
    if type(j[0]) == str or type(j[1]) == str:
      print(j[0]+" "+j[1])
    else:
      print("%3d  %d" % j[0],j[1])

f.close()
```

**0_2. 基本的寫入與讀取**(使用Dict)

```
from bs4 import BeautifulSoup as bs
from google.colab import files
import requests, csv

# 年份
year = []
# 人口數
person = []

url = "https://www.daxi-hro.tycg.gov.tw/home.jsp?id=25&parentpath=0,21,22"

content = requests.get(url).text
parse = bs(content, "html.parser")

data1 = parse.find_all("tbody")
data1 = data1[1]
rows = data1.find_all("tr")
for row in rows:
    cols = row.find_all("td")
    #print(cols)
    if len(cols) > 0 and cols[1].text != '─':
        str1 = cols[0].text
        str1 = str1[0:-1]
        year.append(int(str1))
        #print(year)
        person.append(int(cols[1].text))
        #print(person)

print("桃園市大溪區歷年戶數")
print("=================")

#i = len(year)-1
#while i > -1:
    #print("%3d年: %d戶"%(year[i],person[i]))
    #i -= 1

# 建立csv檔案
with open('桃園市大溪區歷年戶數.csv','w',newline='',encoding='big5') as f :
  # 定義欄位
  fieldnames = ['年份','戶數']
  # 使用DictWriter
  writer = csv.DictWriter(f, fieldnames = fieldnames)
  # 寫入第一位欄位名稱
  writer.writeheader()

  # 寫入資料
  for j in range(len(year)):
    writer.writerow({'年份':year[j],'戶數':person[j]})

f.close()

# 下載檔案
files.download('桃園市大溪區歷年戶數.csv') 

# 讀取csv檔案並輸出
with open('桃園市大溪區歷年戶數.csv','r',newline='',encoding='big5') as f :
  # 使用DictReader
  rows = csv.DictReader(f)
  for j in rows:
    print(j['年份'],j['戶數'])

f.close()
```

**1.使用writerows將多筆資料直接寫入**

```
import csv
from google.colab import files

list1 = [[1, 2, 3], [7, 8, 9], [10, 11, 12]]

with open('test1.csv','w',newline='',encoding='utf-8') as f:
  writer = csv.writer(f)
  writer.writerows(list1)

f.close()

with open('test1.csv','r',newline='',encoding='utf-8') as f :
  reader = csv.reader(f)
  for j in reader:
    print(j)

f.close()

files.download('test1.csv')
```

**2. Dialect** 編碼風格

```
import csv
from google.colab import files

csv.register_dialect("hashes",delimiter="#") # hashes為風格名稱,delimiter為分隔符號

with open('test2.csv','w',newline='',encoding='utf-8') as f:
  writer = csv.writer(f,dialect = 'hashes')
  writer.writerow(('pens',4))
  writer.writerow(('plates',2))
  writer.writerow(('bottles',4))
  writer.writerow(('cups',1))

f.close()

with open('test2.csv','r',newline='',encoding='utf-8') as f: # newline='',可正確解析換行
  reader = csv.reader(f,dialect = 'hashes')
  for j in reader:
    print(j)

f.close()

files.download('test2.csv')
```

```
import csv
from google.colab import files

csv.register_dialect("hashes",delimiter="#") # hashes為風格名稱,delimiter為分隔符號

with open('test2.csv','w',newline='',encoding='utf-8') as f:
  writer = csv.writer(f,dialect = 'hashes')
  writer.writerow(('pens',4))
  writer.writerow(('plates',2))
  writer.writerow(('bottles',4))
  writer.writerow(('cups',1))

f.close()

print(csv.list_dialects())
```
