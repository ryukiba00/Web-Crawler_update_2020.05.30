```
import pandas
from  google.colab import files

# products.csv
products = {"type":["居家","居家","娛樂","娛樂","科技","科技"],
            "name":["家樂福","大潤發","家樂福","全聯超商","大潤發","家樂福"],
            "price":[11.42, 23.50, 19.99, 15.95, 55.75, 111.55]}

df = pandas.DataFrame(products, columns = ["type","name","price"])

df.to_csv("products.csv", index = False, encoding = "big5")

# products2.csv
products = {"type":["居家","娛樂","科技"],
            "name":["家樂福","家樂福","全聯超商"],
            "price":[14.20, 99.90, 66.25]}

df = pandas.DataFrame(products, columns = ["type","name","price"])

df.to_csv("products2.csv", index = False, encoding = "big5")

# type.csv
products = {"type":["居家","娛樂","科技","科學"],
            "num":[25, 75, 15, 10]}

df = pandas.DataFrame(products, columns = ["type","num"])

df.to_csv("type.csv", index = False, encoding = "big5")
#files.download("products.csv")
```

```
import pandas

df = pandas.read_csv("products.csv", encoding = "big5")

# Pandas doesn't allow columns to be created via a new attribute name - see https://pandas.pydata.org/pandas-docs/stable/indexing.html#attribute-access
#df.colums = ["type", "name", "price"]
ordinals = ["A", "B", "C", "D", "E", "F"]

df.index = ordinals
print(df)
```

# **01. 更新**

**1.更新value**

```
df.loc[ordinals[0], "price"] = 21.6 # loc[], 基於行標籤和列標籤（x_label、y_label）進行索引
df.iloc[1, 2] = 46.3  # iloc[], 基於行索引和列索引（index，columns） 都是從 0 開始
print(df.head(2))
```

**2.更新單筆Record**

```
s = ["居家","家樂福", 30.4]

df.loc[ordinals[1]] = s
print(df.head(3))
# B   居家   大潤發   23.50 => B   居家  家樂福  30.40
```

**3. 更新整個欄位值**


```
df.loc[:, "price"] = [23.4, 56.7, 12.1, 90.5, 11.2, 34.1]
print(df.head()) # head()不填值, 預設為5筆
```

**4_1. 更新整個DF**

```
import random

df = pandas.DataFrame([random.sample(range(0, 1000), 3), random.sample(range(0, 1000), 3)]) # random.sample(??, k), 從特定字串或範圍隨機挑選k個元素
                                                                                            # 這邊的格式,[ class1 , class2]為2筆record

print(df)
```

**4_2. 更新整個DF**

```
print(df[df > 500]) # NaN,為Null
print()
df[ df > 500] = df - 100
print(df)
```

# **02. 刪除**

**5. 刪除value**

```
df.loc[ordinals[0], "price"] = None # None, 俗稱Missing Value
df.iloc[1, 2] = None

print(df.head(3))
```

**6. 刪除Record**

```
#df2 = df.drop("D")
df2 = df.drop(["B","D"]) # df.drop(), 刪除record, 如多筆則為df.drop([]), 無法重複刪除
print(df2.head())
print()

df.drop(df.index[[2, 3]], inplace=True) # 少一個[]會變array, 多1個[]則是指定多筆index
print(df.head())
```

**7. 刪除Column(列)**

```
df2 = df.drop(["price"], axis = 1) # axis = 0, 為記錄;axis = 1, 指定為欄位, 可以重複刪除
#df2 = df.drop("price", axis = 1) # "price"有無[], 似乎不受影響
print(df2.head(3))
```

# **03. 新增資料**

**8. 新增Record**

```
df.loc["G"] = ["科學","全聯超商", 28.5]
print(df.tail(3))
print()

s = pandas.Series({"type":"科學","name":"大潤發","price":79.2})
df2 = df.append(s, ignore_index = True)
print(df2.tail(3))
```

**9. 新增欄位**

```
df["sales"] = [124.5, 222.7, 156.7, 435.6, 333.7, 259.8]
print(df.head())
print()

df.loc[:, "city"] = ["台北", "新竹", "台北", "台中", "新北", "高雄"]
print(df.head())
```

# **04. 連接和合併DF**

**10. 建立空的和複製DF**

```
columns = ["type","name","price"]
df_empty = pandas.DataFrame(None, index = ordinals, columns = columns)
#Eprint(df_empty) # 會出現格式跟products一樣的DF, 似乎因為None, 裡面的Record全部變成NaN
#print()

df_copy = df.copy() # df.copy(), 複製df
print(df_copy)
```

**11. 連接多個DF**

```
df2 = pandas.read_csv("products2.csv", encoding = "big5")

df2.index = ["A","B","C"]
df2.columns = columns
#print(df2)
#print()

# 使用pandas.concat([df1, df2]), 似乎也會將重複的index一起複製
df3 = pandas.concat([df, df2]) 
print(df3)
print()

df4 = pandas.concat([df, df2], ignore_index = True)
print(df4)
```

**12. 合併2個DF(類似Sql左右join)**

```
df2 = pandas.read_csv("type.csv", encoding = "big5")

df2.index = ["A","B","C","D"]
df2.columns = ["type","num"]
#print(df2)
#print()

# 使用pandas.merge()
df3 = pandas.merge(df, df2) # 如果沒有符合的欄位,則不會新增紀錄
print(df3)
print()

df4 = pandas.merge(df2, df)
print(df4)
print()

df5 = pandas.merge(df2, df, how = "left") # 多增加參數how = "left or reight or inner"類似Sql的left,right,inner join
print(df5)
```
