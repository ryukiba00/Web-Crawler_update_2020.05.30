```
import pandas
from  google.colab import files

# products.csv
products = {"type":["居家","居家","娛樂","娛樂","科技","科技"],
            "name":["家樂福","大潤發","家樂福","全聯超商","大潤發","家樂福"],
            "price":[11.42, 23.50, 19.99, 15.95, 55.75, 111.55]}

df = pandas.DataFrame(products, columns = ["type","name","price"])

df.to_csv("products.csv", index = False, encoding = "big5")

# products2.csv
products = {"type":["居家","娛樂","科技"],
            
            "name":["家樂福","家樂福","全聯超商"],
            "price":[14.20, 99.90, 66.25]}

df = pandas.DataFrame(products, columns = ["type","name","price"])

df.to_csv("products2.csv", index = False, encoding = "big5")

# type.csv
products = {"type":["居家","娛樂","科技","科學"],
            "num":[25, 75, 15, 10]}

df = pandas.DataFrame(products, columns = ["type","num"])

df.to_csv("type.csv", index = False, encoding = "big5")

# products3.csv
products = {"type":["居家","居家","娛樂","娛樂","科技","科技","居家","娛樂","科技"],
            "name":["家樂福","大潤發","家樂福","全聯超商","大潤發","家樂福","家樂福","家樂福","全聯超商"],
            "price":[11.42, 23.50, 19.99, 15.95, 55.75, 111.55,14.20,99.90,66.25]}

df = pandas.DataFrame(products, index = ["A","B","C","D","E","F","G","H","I"] ,columns = ["type","name","price"])

df.to_csv("products3.csv", encoding = "big5")
df.to_html("products3.html", encoding = "big5")

#files.download("products3.csv")
#files.download("products3.html")
```

# **1. 群組(講簡單點就是Sql的Group By)**

**1. 使用群組來計算加總及平均**

```
df = pandas.read_csv("products3.csv", encoding = "big5")

# 很像sql語法, 將type進行分類, 之後再各自加總
print(df.groupby("type").sum())
print()

print(df.groupby(["type","name"]).mean()) 

df2 = df.groupby(["type","name"]).mean()
df2.to_html("products3.html", encoding = "big5")

files.download("products3.html")
```

# **2. 樞紐分析表(Excel的樞紐分析功能)**

**1. 將DF建立成樞紐分析表**

```
df = pandas.read_csv("products.csv", encoding = "big5")

pivot_products = df.pivot_table(index = "type", columns = "name", values = "price")

print(pivot_products)

pivot_products.to_html("products.html", encoding = "big5")
files.download("products.html")
```

# **3. 函數**

**1. 套用函數**

```
def double(x):
  return x*2

df2 = df["price"].apply(double)
print(df2)
```

**2. 套用Lambda運算式**

```
df2 = df["price"].apply(lambda x : x*2)
print(df2)
```

# **4. DF的統計函數**

**1. Pandas內建的describe()函數**

```
print(df["price"].describe())
```

**2. 其他統計函數(count())**

```
# 進行非NaN值計數
print(df["price"].count())
```

**3. 其他統計函數(mode())**

```
# 取眾數
print(df["price"].mode())
```

**4. 其他統計函數(median())**

```
# 取中位數
print(df["price"].median())
```

**5. 其他統計函數(quantile())**

```
# 取四分位數
print(df["price"].quantile()) # 預設為0.5
print(df["price"].quantile(q=0.25))
print(df["price"].quantile(q=0.5))
print(df["price"].quantile(q=0.75))
```

**6. 其他統計函數(mean())**

```
print(df["price"].mean())
```

**7. 其他統計函數(max())**

```
print(df["price"].max())
```

**8. 其他統計函數(min())**

```
print(df["price"].min())
```

**9. 其他統計函數(sum())**

```
print(df["price"].sum())
```

**10. 其他統計函數(var())**

```
# 取變異數
print(df["price"].var())
```

**11. 其他統計函數(std())**

```
# 取標準差
print(df["price"].std())
```

**12. 其他統計函數(cov())**

```
# 取共變異數
print(df.cov())
```

**13. 其他統計函數(corr())**

```
# 取相關係數
print(df.corr())
```

**14. 其他統計函數(cumsum())**

```
# 取累積總和
print(df["price"].cumsum())
```

**15. 其他統計函數(cumprod())**

```
# 取累積乘積
print(df["price"].cumprod())
```
