**3_1.更改DF欄位名稱** (使用rename())

```
import pandas

dic = {
    "col 1" : [1, 2, 3],
    "col 2" : [10, 20, 30],
    "col 3" : list('abc'),
    "col 4" : ['a','b','c'],
    "col 5" : pandas.Series(range(3))
}

df = pandas.DataFrame(dic)

# 更改動作，使用reanme()
rename_dic = {"col 1" : 'x', 'col 2': '10x'}

df = df.rename(rename_dic, axis=1) # 在 pandas 裡大部分函式預設處理的軸為列（row）：以 axis=0 表示；而將 axis 設置為 1 則代表你想以行（column）為單位套用該函式。

df 
```

**10_2.建立自訂索引的Series物件, 其他操作**

```
import pandas

fruits = ["蘋果", "橘子", "梨子", "櫻桃"]
quantities = [15, 33, 45, 55]

s = pandas.Series(quantities, index = fruits)
# 指定index,取出資料
print("橘子= ",s["橘子"])
print()
# 指定多個index,取出多個資料,使用兩個[]
print(s[["橘子","梨子","櫻桃"]])
print()
# Series可執行運算
s = (s + 2) * 3
print(s)
```

**12_1. 建立自訂索引的DF物件, 其他操作**

```
import pandas

products = {"分類":["居家","居家","娛樂","娛樂","科技","科技"],
            "商店":["家樂福","大潤發","家樂福","全聯超商","大潤發","家樂福"],
            "價格":[11.42, 23.50, 19.99, 15.95, 55.75, 111.55]}

ordinals = ["A","B","C","D","E","F"]

df = pandas.DataFrame(products, index = ordinals)
print(df)
print()
print("***不同的改變index的方式***")
df2 = pandas.DataFrame(products)
df2.index = ordinals
print(df)
print()
print("***重新指定DF的欄位順序***")
df = pandas.DataFrame(products, columns = ["價格","分類","商店"], index = ordinals)
print(df)
print()
print("***使用存在的column做為index***")
df = pandas.DataFrame(products, columns = ["商店","價格"], index = products["分類"])
print(df)
print()
print("***轉置***")
print(df.T)
```
