```
df = pandas.read_csv("missing_data.csv")
print(df)
```

# **01. 處理遺漏值**

**01_1. 顯示遺漏值的資訊**

```
df.info() 
```

**01_2. 刪除屬於NaN的Record**

```
# 使用dropna(), 預設為刪除全部, 可以增加how參數
df1 = df.dropna()
print(df1)
print()

# how ="any", 刪除任何含NaN的Record
df2 = df.dropna(how = "any")
print(df2)
print()

# how = "all", 刪除所有欄位都是NaN的Record
df3 = df.dropna(how = "all")
print(df3)
print()

# subset屬性指定某些欄位, 只要有NaN就會刪除
df4 = df.dropna(subset = ["COL_B","COL_C"])
print(df4)
```

**01_3. 填補遺漏值**

```
# 使用fillna()，並指定參數value
df1 = df.fillna(value = 1)
print(df1)
print()

# 也可以加入mean()其他統計函數
df["COL_B"] = df["COL_B"].fillna(df["COL_B"].mean())
print(df)
print()

df["COL_C"] = df["COL_C"].fillna(df["COL_C"].median())
print(df)
print()
```

# **02. 處理重覆資料**

**02_0. 讀取檔案**

```
df = pandas.read_csv("duplicated_data.csv")
print(df)
```

**02_1. 刪除重覆記錄**

```
# 使用drop_duplicates()刪除所有重覆的Record, 但不包含第一筆Record
df1 = df.drop_duplicates()
print(df1)
```

**02_2. 刪除重覆欄位值**

```
# 指定欄位, 預設會留第一個重覆record
df1 = df.drop_duplicates("COL_B")
print(df1)
print()

# 添加參數keep, 指定要保留第n個重覆record, 其餘刪除
df2 = df.drop_duplicates("COL_B", keep = "last")
print(df2)
print()

# 如果參數keep = False, 所有重覆record的不會保留, 全部刪除
df3 = df.drop_duplicates("COL_B", keep = False)
print(df3)
```

# **03. 轉換分類資料**

```
f = {"Gender":["male","female","not specified","male","female","female"],
     "Size":["XL","M","XXL","L","S","XS"],
    "Price":[800, 400, 300, 500, 700, 850]}
df = pandas.DataFrame(f)
print(df)
print()

size_mapping = {"XXL":5,
                "XL":4,
                "L":3,
                "M":2,
                "S":1,
                "XS":0}

df["Size"] = df["Size"].map(size_mapping)
print(df)
```
