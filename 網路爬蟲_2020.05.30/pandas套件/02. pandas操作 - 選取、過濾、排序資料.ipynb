**1_2. 選取單一欄位或多個欄位(進階)**

```
import pandas
from google.colab import files

df = pandas.read_csv("products.csv", encoding = "big5")

ordinals = ["A", "B", "C", "D", "E", "F"]
df.index = ordinals

# loc(), 使用loc指定索引標籤和欄位標籤
print(df.loc["C":"E",["name","price"]])
print()
print(df.loc["C",["name","price"]])
#df2 = df.loc["C":"E",["name","price"]]
#df2 = df.loc["C",["name","price"]]
print()

print(df.loc[ordinals[0],"name"])
print()
print(type(df.loc[ordinals[0],"name"]))
print()
print(df.loc["A", "price"])
print()
print(type(df.loc["A","price"]))
print()

# iloc(), 使用iloc索引器,針對位置索引
print(df.iloc[3])
print()
# 可進行切割
print(df.iloc[3:5, 1:3]) # 跟loc()一樣, [3:5]為自訂標籤的索引 (3~4), [1:3]為欄位標籤的索引, (name, price)
print()
print(df.iloc[1:3, :])
print()
print(df.iloc[:, 1:3])
print()

print(df.iloc[[1, 2, 4], [0, 2]])
print(df.iloc[1,1])
print(df.iat[1,1])

#df2.to_html("products.html")
#files.download("products.html")
```

**2. 過濾資料**

```
import pandas
from google.colab import files

df = pandas.read_csv("products.csv", encoding = "big5")

ordinals = ["A", "B", "C", "D", "E", "F"]
df.index = ordinals

print(df.price > 20)
print()
print(df[df.price > 20])

# 使用isin檢查是否在df裡, 並return 一整個record
print(df[df["type"].isin(["科技","居家"])])
print()

print(df[(df.price > 15) & (df.price < 25)]) # 不能使用and 跟or 
print()

# 新增record
df.loc["G"] = ["科學", "全連超商", 28.5]
print(df[df["type"].str.startswith("科")])

#df2.to_html("products.html")
#files.download("products.html")
```

**3.排序資料**

```
import pandas
from google.colab import files

df = pandas.read_csv("products.csv", encoding = "big5")

ordinals = ["A", "B", "C", "D", "E", "F"]
df.index = ordinals

# 指定索引欄位排序
df2 = df.set_index("price")
#print(df2)
df2.sort_index(ascending=False, inplace= True)
print(df2)
print()

# 指定欄位值排序 
df2 = df.sort_values("price", ascending=False)
print(df2)
print()

df.sort_values(["type","price"], inplace = True)
print(df)
```
